# Types of convergence

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the number of iterations
iterations = np.arange(1, 21)

# Simulated error decay for different types of convergence
linear = 0.5 ** iterations  # Linear: constant factor reduction
superlinear = 0.9 ** (iterations ** 1.5)  # Superlinear: faster than linear
quadratic = 1 / (iterations ** 2)  # Quadratic: error squared each step
sublinear = 1 / (iterations + 1)  # Sublinear: slow decay

# Plotting the convergence behavior
plt.figure(figsize=(10, 6))
plt.plot(iterations, linear, 'r-', linewidth=2, label='Linear Convergence')
plt.plot(iterations, superlinear, 'b-', linewidth=2, label='Superlinear Convergence')
plt.plot(iterations, quadratic, 'g-', linewidth=2, label='Quadratic Convergence')
plt.plot(iterations, sublinear, 'orange', linewidth=2, label='Sublinear Convergence')

plt.yscale('log')  # Log scale to visualize rates clearly
plt.xlabel('Iteration')
plt.ylabel('Error (log scale)')
plt.title('Types of Convergence (Simulated Error Decay)')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

# Convergence with radius

```python
import numpy as np
import matplotlib.pyplot as plt

# Define iterations
iterations = np.arange(1, 21)

# Simulated error sequences
linear = 0.5 ** iterations  # Linear: reduces error by constant ratio
superlinear = 0.9 ** (iterations ** 1.5)  # Superlinear: faster drop-off
quadratic = 1 / (iterations ** 2)  # Quadratic: very fast near convergence
sublinear = 1 / (iterations + 1)  # Sublinear: slow decay

# Define a convergence radius (ε-ball)
convergence_radius = 1e-2  # This threshold represents convergence

# Create the plot
plt.figure(figsize=(10, 6))
plt.plot(iterations, linear, 'r-', linewidth=2, label='Linear Convergence')
plt.plot(iterations, superlinear, 'b-', linewidth=2, label='Superlinear Convergence')
plt.plot(iterations, quadratic, 'g-', linewidth=2, label='Quadratic Convergence')
plt.plot(iterations, sublinear, 'orange', linewidth=2, label='Sublinear Convergence')

# Add convergence radius line
plt.axhline(y=convergence_radius, color='black', linestyle='--', linewidth=2,
            label='Convergence Radius (ε = 1e-2)')

# Log scale for better visibility
plt.yscale('log')
plt.xlabel('Iteration')
plt.ylabel('Error (log scale)')
plt.title('Convergence Types with Convergence Radius')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
