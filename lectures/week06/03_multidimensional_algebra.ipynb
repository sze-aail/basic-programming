# Laplacian operator

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the scalar field function
def f(x, y):
    return np.sin(x) * np.cos(y)

# Generate the grid
x = np.linspace(-2 * np.pi, 2 * np.pi, 200)
y = np.linspace(-2 * np.pi, 2 * np.pi, 200)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Compute grid spacing
dx = x[1] - x[0]
dy = y[1] - y[0]

# Compute the Laplacian ∆f = ∂²f/∂x² + ∂²f/∂y²
d2f_dx2 = (np.roll(Z, -1, axis=1) - 2 * Z + np.roll(Z, 1, axis=1)) / dx**2
d2f_dy2 = (np.roll(Z, -1, axis=0) - 2 * Z + np.roll(Z, 1, axis=0)) / dy**2
laplacian = d2f_dx2 + d2f_dy2

# Plot both the function and Laplacian on the same 3D surface
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the original function as a wireframe with transparency
ax.plot_wireframe(X, Y, Z, color='gray', alpha=0.9, linewidth=0.5, label='f(x, y)')

# Overlay the Laplacian as a colored surface
surf = ax.plot_surface(X, Y, laplacian, cmap='coolwarm', edgecolor='none', alpha=0.5)

# Formatting
ax.set_title('Combined Visualization: f(x, y) and Laplacian ∆f(x, y)')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
fig.colorbar(surf, ax=ax, shrink=0.5, aspect=10, label='Laplacian ∆f')
plt.tight_layout()
plt.show()
```

## Curl operator
The curl operator is a vector operator that describes the rotation of a vector field in three-dimensional space. It is defined as:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a 2D vector field F = [F1, F2]
def F(x, y):
    F1 = -np.sin(y)   # x-component
    F2 = np.sin(x)    # y-component
    return F1, F2

# Generate the grid
x = np.linspace(-2 * np.pi, 2 * np.pi, 40)
y = np.linspace(-2 * np.pi, 2 * np.pi, 40)
X, Y = np.meshgrid(x, y)

# Compute the vector field on the grid
F1, F2 = F(X, Y)

# Grid spacing
dx = x[1] - x[0]
dy = y[1] - y[0]

# Compute partial derivatives for curl in 2D (only z-component exists)
dF2_dx = (np.roll(F2, -1, axis=1) - np.roll(F2, 1, axis=1)) / (2 * dx)
dF1_dy = (np.roll(F1, -1, axis=0) - np.roll(F1, 1, axis=0)) / (2 * dy)

# Curl z-component (∂F2/∂x - ∂F1/∂y)
curl_z = dF2_dx - dF1_dy

# Create the plot
plt.figure(figsize=(10, 8))

# Quiver plot of the vector field
plt.quiver(X, Y, F1, F2, color='black', scale=30, width=0.003)

# Overlay the curl as a color-filled contour plot
plt.contourf(X, Y, curl_z, levels=100, cmap='coolwarm', alpha=0.8)
plt.colorbar(label='Curl z-component ∇×F')

# Formatting
plt.title('2D Vector Field and Its Curl (z-component)')
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Divergence operator
The divergence operator measures the magnitude of a vector field's source or sink at a given point. It is defined as:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the 2D vector field F = [F1, F2]
def F(x, y):
    F1 = np.sin(x) * np.cos(y)
    F2 = -np.cos(x) * np.sin(y)
    return F1, F2

# Generate the grid
x = np.linspace(-2 * np.pi, 2 * np.pi, 40)
y = np.linspace(-2 * np.pi, 2 * np.pi, 40)
X, Y = np.meshgrid(x, y)

# Evaluate the vector field
F1, F2 = F(X, Y)

# Grid spacing
dx = x[1] - x[0]
dy = y[1] - y[0]

# Compute partial derivatives for divergence
dF1_dx = (np.roll(F1, -1, axis=1) - np.roll(F1, 1, axis=1)) / (2 * dx)
dF2_dy = (np.roll(F2, -1, axis=0) - np.roll(F2, 1, axis=0)) / (2 * dy)

# Divergence: ∇·F = ∂F1/∂x + ∂F2/∂y
div_F = dF1_dx + dF2_dy

# Create the plot
plt.figure(figsize=(10, 8))

# Quiver plot for the vector field
plt.quiver(X, Y, F1, F2, color='black', scale=30, width=0.003)

# Divergence heatmap
plt.contourf(X, Y, div_F, levels=100, cmap='coolwarm', alpha=0.8)
plt.colorbar(label='Divergence ∇·F')

# Add title with the function definition
plt.title(r'2D Vector Field $\vec{F}(x, y) = [\sin(x)\cos(y),\ -\cos(x)\sin(y)]$ and Its Divergence')

# Labels and layout
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## Gradient operator
The gradient operator measures the rate and direction of change in a scalar field. It is defined as:

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the scalar field
def f(x, y):
    return np.sin(x) * np.cos(y)

# Generate a 2D grid
x = np.linspace(-2 * np.pi, 2 * np.pi, 40)
y = np.linspace(-2 * np.pi, 2 * np.pi, 40)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Calculate grid spacing
dx = x[1] - x[0]
dy = y[1] - y[0]

# Compute partial derivatives (gradient components) using central differences
df_dx = (np.roll(Z, -1, axis=1) - np.roll(Z, 1, axis=1)) / (2 * dx)
df_dy = (np.roll(Z, -1, axis=0) - np.roll(Z, 1, axis=0)) / (2 * dy)

# Plotting the scalar field and its gradient vectors
plt.figure(figsize=(10, 8))

# Scalar field as contour background
plt.contourf(X, Y, Z, levels=100, cmap='viridis', alpha=0.8)
plt.colorbar(label='Scalar Field f(x, y)')

# Quiver plot: gradient vectors
plt.quiver(X, Y, df_dx, df_dy, color='white', scale=50)

# Title includes the definition of the scalar field
plt.title(r'Gradient of Scalar Field $f(x, y) = \sin(x)\cos(y)$')

# Axes and layout
plt.xlabel('x')
plt.ylabel('y')
plt.axis('equal')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### In 3D
The gradient operator in three dimensions is defined as:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the scalar field
def f(x, y):
    return np.sin(x) * np.cos(y)

# Create the grid
x = np.linspace(-2 * np.pi, 2 * np.pi, 30)
y = np.linspace(-2 * np.pi, 2 * np.pi, 30)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Grid spacing
dx = x[1] - x[0]
dy = y[1] - y[0]

# Compute gradient using central differences
df_dx = (np.roll(Z, -1, axis=1) - np.roll(Z, 1, axis=1)) / (2 * dx)
df_dy = (np.roll(Z, -1, axis=0) - np.roll(Z, 1, axis=0)) / (2 * dy)

# Normalize gradient for arrow directions
norm = np.sqrt(df_dx**2 + df_dy**2)
norm[norm == 0] = 1
U = df_dx / norm
V = df_dy / norm
W = np.zeros_like(U)

# Create the 3D plot
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# Plot scalar surface
ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='k', linewidth=0.2, alpha=0.8)

# Plot gradient vectors on the surface
ax.quiver(X, Y, Z, U, V, W, length=0.5, color='red', normalize=True)

# Labels and title
ax.set_title(r'3D Gradient of Scalar Field $f(x, y) = \sin(x)\cos(y)$')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('f(x, y)')
plt.tight_layout()
plt.show()
```
