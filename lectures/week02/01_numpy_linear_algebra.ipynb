## Dot product of two vectors
Visualize the dot product of two vectors using NumPy with quiver plot

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d.art3d import Poly3DCollection

import seaborn as sns
import matplotlib.animation as animation
from matplotlib.patches import Ellipse
```

```python
# Define two vectors
v1 = np.array([3, 2])  # First vector
v2 = np.array([2, 3])  # Second vector

# Compute the dot product
dot_product = np.dot(v1, v2)

# Compute the projection of v1 onto v2
v2_unit = v2 / np.linalg.norm(v2)  # Unit vector of v2
proj_v1_on_v2 = np.dot(v1, v2_unit) * v2_unit  # Projection formula

# Create a figure
fig, ax = plt.subplots(figsize=(6, 6))

# Plot vectors
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='blue', label=r"v1")
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='red', label=r"v2")
ax.quiver(0, 0, proj_v1_on_v2[0], proj_v1_on_v2[1], angles='xy', scale_units='xy', scale=1, color='green', linestyle='dashed', label=r"v2_proj_v_1")

# Add annotations
ax.text(v1[0], v1[1], r"v1", fontsize=12, verticalalignment='bottom', horizontalalignment='right', color='blue')
ax.text(v2[0], v2[1], r"v2", fontsize=12, verticalalignment='bottom', horizontalalignment='left', color='red')
ax.text(proj_v1_on_v2[0], proj_v1_on_v2[1], r"v2_proj_v1", fontsize=12, verticalalignment='top', horizontalalignment='right', color='green')

# Add text for the dot product
ax.text(0.5, -0.5, f"Dot Product: {dot_product}", fontsize=14, bbox=dict(facecolor='white', alpha=0.6))

# Set limits
ax.set_xlim(-1, 5)
ax.set_ylim(-1, 5)

# Grid and axis labels
ax.axhline(0, color='black', linewidth=1)
ax.axvline(0, color='black', linewidth=1)
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_title("Dot Product Visualization")

# Show legend

# Show plot
plt.grid(True, linestyle="--", alpha=0.5)
plt.show()
```

## Norms of vectors
Visualize the norms of vectors using NumPy with quiver plot

```python


def plot_norms(p_values=[1, 2, np.inf]):
    """
    Function to visualize different norms in 2D for given p-values.

    Parameters:
    - p_values: List of norm orders (e.g., [1, 2, np.inf])
    """
    fig, ax = plt.subplots(len(p_values), 1, figsize=(6, 12))

    # Generate a grid for visualizing norms
    x = np.linspace(-4, 4, 100)
    y = np.linspace(-4, 4, 100)
    X, Y = np.meshgrid(x, y)

    # Compute level sets for each norm based on p-values
    norm_shapes = {}
    for p in p_values:
        if p == 1:
            norm_shapes["Manhattan Norm (L1)"] = np.abs(X) + np.abs(Y)
        elif p == 2:
            norm_shapes["Euclidean Norm (L2)"] = np.sqrt(X**2 + Y**2)
        elif p == np.inf:
            norm_shapes["Infinity Norm (L∞)"] = np.maximum(np.abs(X), np.abs(Y))
        else:
            norm_shapes[f"Generalized Norm (L{p})"] = (np.abs(X) ** p + np.abs(Y) ** p) ** (1 / p)

    # Plot each norm visualization
    for i, (title, norm_data) in enumerate(norm_shapes.items()):
        cs = ax[i].contour(X, Y, norm_data, levels=[1, 2, 3], colors=['blue', 'red', 'green'])
        ax[i].clabel(cs, inline=True, fontsize=10)
        ax[i].set_title(title)
        ax[i].axhline(0, color='black', linewidth=1)
        ax[i].axvline(0, color='black', linewidth=1)
        ax[i].grid(True, linestyle="--", alpha=0.5)

    plt.tight_layout()
    plt.show()

# Example usage: Visualize L1, L2, and L∞ norms, and an additional p=3 norm
plot_norms(p_values=[1,2,np.inf,3])
```

### Matrix multiplication
Visualize matrix multiplication using NumPy

```python
# Function to visualize a matrix using a heatmap
def plot_matrix(ax, matrix, title, cmap="coolwarm"):
    sns.heatmap(matrix, annot=True, fmt=".1f", cmap=cmap, linewidths=0.5, ax=ax, cbar=False)
    ax.set_title(title, fontsize=14)
    ax.set_xticks([])
    ax.set_yticks([])

# Generate two random matrices (3x3)
np.random.seed(42)
A = np.random.randint(-5, 5, (3, 3))
B = np.random.randint(-5, 5, (3, 3))

# Define operations
scalar = 2
A_plus_B = A + B
A_times_B = np.dot(A, B)
A_transpose = A.T
A_inverse = np.linalg.inv(A) if np.linalg.det(A) != 0 else None  # Check if A is invertible

# Create figure with subplots
fig, axes = plt.subplots(3, 3, figsize=(12, 12))

# Plot original matrices
plot_matrix(axes[0, 0], A, "Matrix A")
plot_matrix(axes[0, 1], B, "Matrix B")
plot_matrix(axes[0, 2], scalar * A, f"Scalar Multiplication (2 * A)")

# Plot results of operations
plot_matrix(axes[1, 0], A_plus_B, "Matrix Addition (A + B)")
plot_matrix(axes[1, 1], A_times_B, "Matrix Multiplication (A * B)")
plot_matrix(axes[1, 2], A_transpose, "Transpose of A (Aᵀ)")

# Plot inverse if it exists
if A_inverse is not None:
    plot_matrix(axes[2, 1], A_inverse, "Inverse of A (A⁻¹)")
    axes[2, 0].axis("off")
    axes[2, 2].axis("off")
else:
    axes[2, 0].text(0.5, 0.5, "A is Not Invertible", ha="center", va="center", fontsize=14, color="red")
    axes[2, 0].set_title("Inverse of A (Not Possible)")
    axes[2, 1].axis("off")
    axes[2, 2].axis("off")

# Adjust layout and show the figure
plt.tight_layout()
plt.show()
```

## Gaussian elimination
Visualize Gaussian elimination using NumPy

```python

def gaussian_elimination(A, b):
    """
    Performs Gaussian elimination step by step and prints each transformation.
    Solves Ax = b for x using forward elimination and back-substitution.
    """
    A = A.astype(float)  # Ensure float division
    b = b.astype(float)
    n = len(A)

    # Augment matrix A with vector b
    Ab = np.hstack([A, b.reshape(-1, 1)])
    print("Initial Augmented Matrix:")
    print(Ab, "\n")

    # Forward elimination
    for i in range(n):
        # Partial Pivoting: Swap rows if needed
        max_row = np.argmax(abs(Ab[i:, i])) + i
        if i != max_row:
            Ab[[i, max_row]] = Ab[[max_row, i]]
            print(f"Swapped row {i} with row {max_row} for pivoting:")
            print(Ab, "\n")

        # Make diagonal element 1 (Normalization)
        pivot = Ab[i, i]
        if pivot == 0:
            print("Singular matrix detected. Cannot proceed.")
            return None

        Ab[i] = Ab[i] / pivot
        print(f"Row {i} normalized (pivot = {pivot}):")
        print(Ab, "\n")

        # Make below elements in column zero
        for j in range(i + 1, n):
            factor = Ab[j, i]
            Ab[j] -= factor * Ab[i]
            print(f"Eliminating row {j} using row {i} (factor = {factor}):")
            print(Ab, "\n")

    # Back-substitution
    x = np.zeros(n)
    for i in range(n - 1, -1, -1):
        x[i] = Ab[i, -1] - np.sum(Ab[i, i + 1:n] * x[i + 1:n])
        print(f"Back-substituting for x[{i}]: {x[i]}")

    print("\nFinal solution vector x:")
    print(x)
    return Ab, x

# Example system Ax = b
A = np.array([[2, -1, 1], [1, 3, 2], [1, -1, 2]])
b = np.array([8, 13, 3])

# Solve and print step-by-step
Ab, solution = gaussian_elimination(A, b)

print("\nFinal Augmented Matrix:\n", Ab)
```

Visualizing the solution of a system of linear equations and the corresponding planes

```python

def plot_planes(A, b, title="Visualization of Planes"):
    """
    Plots the three planes defined by Ax = b in 3D.
    """
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')

    # Define grid
    x_vals = np.linspace(-10, 10, 20)
    y_vals = np.linspace(-10, 10, 20)
    X, Y = np.meshgrid(x_vals, y_vals)

    # Plot each plane
    colors = ['cyan', 'magenta', 'yellow']
    for i in range(3):
        print(A[i])
        a0, b0, c0 = A[i]
        d = b[i]
        if c0 == 0:
            continue  # Avoid division by zero for vertical planes
        Z = (d - a0 * X - b0 * Y) / c0
        ax.plot_surface(X, Y, Z, color=colors[i], alpha=0.5, edgecolor='k')

    # Labels and Title
    ax.set_xlabel('X-axis')
    ax.set_ylabel('Y-axis')
    ax.set_zlabel('Z-axis')
    ax.set_title(title)
    plt.show()

# Example system Ax = b
A = np.array([[2, -1, 1], [1, 3, 2], [1, -1, 2]])
b = np.array([8, 13, 3])

# Solve using Gaussian elimination
A0, b0 = gaussian_elimination(A, b)
A0 = A0[:, :-1]  # Remove the last column (b values)


# Plot the original system of planes
plot_planes(A, b, title="Original System of Equations")

# If solution exists, plot the modified system
if solution is not None:
    plot_planes(A0, b0, title="Gaussian Elimination Solution")

```

### Solve engineering problem
Solving an engineering problem using NumPy: optimizing structural loads

```python
# Coefficients Matrix (Force Balance Equations)
A = np.array([
    [1, 1, 1],   # Sum of forces in horizontal direction (ΣF_x = 0)
    [0, 1, -1],  # Vertical equilibrium condition (ΣF_y = 0)
    [2, -1, 1]   # Torque equilibrium (ΣM = 0)
])

# External force vector (applied loads)
b = np.array([100, 50, 75])  # External loads on the structure

# Solve for internal forces using NumPy
forces = np.linalg.solve(A, b)

# Output the computed forces
print("Optimized Internal Forces in the Structure:")
for i, f in enumerate(forces):
    print(f"Force in Member {i+1}: {f:.2f} N")

# Validate solution (A * forces should equal b)
residual = np.dot(A, forces) - b
print("\nValidation (Residuals):", residual)
```

Visualizing the solution of a system of linear equations and the corresponding lines

```python
# Define the truss structure with joints and members
joints = np.array([
    [0, 0],  # Joint 1 (fixed)
    [2, 2],  # Joint 2 (middle)
    [4, 0]   # Joint 3 (fixed)
])

members = [
    (0, 1),  # Member 1 (Joint 1 to Joint 2)
    (1, 2),  # Member 2 (Joint 2 to Joint 3)
    (0, 2)   # Member 3 (Joint 1 to Joint 3)
]

# Coefficients Matrix (Force Balance Equations)
A = np.array([
    [1, 1, 1],   # Sum of forces in horizontal direction (ΣF_x = 0)
    [0, 1, -1],  # Vertical equilibrium condition (ΣF_y = 0)
    [2, -1, 1]   # Torque equilibrium (ΣM = 0)
])

# External force vector (applied loads)
b = np.array([100, 50, 75])  # External loads on the structure

# Solve for internal forces using NumPy
forces = np.linalg.solve(A, b)

# Plot the truss structure
fig, ax = plt.subplots(figsize=(6, 6))

# Plot members with color coding (Red for compression, Blue for tension)
for i, (start, end) in enumerate(members):
    x_vals = [joints[start, 0], joints[end, 0]]
    y_vals = [joints[start, 1], joints[end, 1]]
    color = 'red' if forces[i] < 0 else 'blue'  # Red for compression, Blue for tension
    ax.plot(x_vals, y_vals, color=color, linewidth=3, label=f"Member {i+1}: {forces[i]:.2f} N")

# Plot joints
ax.scatter(joints[:, 0], joints[:, 1], color='black', s=100, label="Joints")

# Annotate joints
for i, (x, y) in enumerate(joints):
    ax.text(x, y, f" {i+1}", fontsize=12, verticalalignment='bottom', horizontalalignment='right', color='black')

# Labels and formatting
ax.set_xlabel("X Position (m)")
ax.set_ylabel("Y Position (m)")
ax.set_title("Truss Structure Visualization with Optimized Forces")
ax.legend()
ax.grid(True, linestyle="--", alpha=0.5)

# Show plot
plt.show()
```

## Visualization of independent and dependent systems of linear equations
Visualizing independent and dependent systems of linear equations

```python

def plot_vectors(vectors, title, colors=None):
    """
    Plots a set of 2D vectors to visualize linear dependence and independence.
    """
    fig, ax = plt.subplots(figsize=(6, 6))

    # Set colors for vectors
    if colors is None:
        colors = ['blue', 'red', 'green']

    # Plot vectors
    for i, v in enumerate(vectors):
        ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color=colors[i], label=f"v{i+1}")

    # Formatting
    ax.set_xlim(-2, 2)
    ax.set_ylim(-2, 2)
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)
    ax.grid(True, linestyle="--", alpha=0.5)
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.set_title(title)
    ax.legend()

    # Show plot
    plt.show()

# Define linearly dependent vectors (same direction or multiples of each other)
dependent_vectors = [
    np.array([1, 1]),
    np.array([2, 2]),  # Multiple of first vector (Linearly Dependent)
    np.array([-1, -1]) # Another multiple of first vector (Still Dependent)
]

# Define linearly independent vectors (spanning the space)
independent_vectors = [
    np.array([1, 0]),  # Basis vector in X direction
    np.array([0, 1]),  # Basis vector in Y direction
    np.array([1, 1])   # Another vector not collinear with the first two
]

# Plot linearly dependent vectors
plot_vectors(dependent_vectors, "Linearly Dependent Vectors", colors=['blue', 'red', 'green'])

# Plot linearly independent vectors
plot_vectors(independent_vectors, "Linearly Independent Vectors", colors=['blue', 'red', 'green'])
```

## Visualize the inverse
Visualizing the inverse of a matrix using NumPy

```python

def visualize_matrix_inverse(A):
    """
    Visualizes the inverse of a matrix using a heatmap.
    """
    # Compute the inverse
    try:
        A_inv = np.linalg.inv(A)
    except np.linalg.LinAlgError:
        print("The matrix is singular and cannot be inverted.")
        return

    # Create a figure
    fig, axes = plt.subplots(2, 1, figsize=(5, 10))

    # Plot the original matrix
    sns.heatmap(A, annot=True, cmap="coolwarm", center=0, ax=axes[0])
    axes[0].set_title("Original Matrix A")

    # Plot the inverse matrix
    sns.heatmap(A_inv, annot=True, cmap="coolwarm", center=0, ax=axes[1])
    axes[1].set_title("Inverse Matrix A⁻¹")

    # Display the plots
    plt.tight_layout()
    plt.show()

# Example 3x3 matrix
A = np.array([
    [2, 3, 1],
    [1, 2, 3],
    [3, 1, 2]
])

# Call the visualization function
visualize_matrix_inverse(A)
```

## Applications of determinants

Visualizing the determinant of a matrix

```python

def visualize_3d_parallelepiped(A):
    """
    Visualizes the determinant of a 3×3 matrix as the volume of a parallelepiped.
    """
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')

    # Define the origin and vectors from the matrix columns
    origin = np.array([0, 0, 0])
    v1, v2, v3 = A[:, 0], A[:, 1], A[:, 2]

    # Compute the vertices of the parallelepiped
    vertices = np.array([
        origin, v1, v1 + v2, v2,  # Bottom face
        v3, v1 + v3, v1 + v2 + v3, v2 + v3  # Top face
    ])

    # Define faces of the parallelepiped
    faces = [
        [vertices[0], vertices[1], vertices[5], vertices[4]],  # Side 1
        [vertices[1], vertices[2], vertices[6], vertices[5]],  # Side 2
        [vertices[2], vertices[3], vertices[7], vertices[6]],  # Side 3
        [vertices[3], vertices[0], vertices[4], vertices[7]],  # Side 4
        [vertices[4], vertices[5], vertices[6], vertices[7]],  # Top face
        [vertices[0], vertices[1], vertices[2], vertices[3]],  # Bottom face
    ]

    # Plot the parallelepiped
    ax.add_collection3d(Poly3DCollection(faces, color="cyan", alpha=0.5, edgecolor='k'))

    # Plot vectors
    ax.quiver(*origin, *v1, color='r', label='v1', linewidth=2)
    ax.quiver(*origin, *v2, color='g', label='v2', linewidth=2)
    ax.quiver(*origin, *v3, color='b', label='v3', linewidth=2)

    # Labels and limits
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.set_zlabel("Z-axis")
    ax.set_title(f"3D Parallelepiped Visualization (Det = {np.linalg.det(A):.2f})")
    ax.legend()
    ax.set_xlim([0, max(A[0]) + 4])
    ax.set_ylim([0, max(A[1]) + 4])
    ax.set_zlim([0, max(A[2]) + 4])

    plt.show()

# Example 3x3 matrix for visualization
A_3D = np.array([
    [2, 1, 3],
    [1, 2, 1],
    [3, 1, 2]
])

visualize_3d_parallelepiped(A_3D)
```

```python

def visualize_grid_transformation(A):
    """
    Visualizes how a 2×2 matrix transformation affects a 2D grid.
    """
    # Define the original grid points
    x = np.linspace(-2, 2, 10)
    y = np.linspace(-2, 2, 10)
    X, Y = np.meshgrid(x, y)
    grid_points = np.vstack([X.ravel(), Y.ravel()])

    # Apply the matrix transformation
    transformed_points = A @ grid_points
    X_transformed = transformed_points[0, :].reshape(X.shape)
    Y_transformed = transformed_points[1, :].reshape(Y.shape)

    # Create the plot
    fig, ax = plt.subplots(figsize=(8, 6))

    # Plot original grid
    ax.plot(X, Y, 'k--', linewidth=0.5, alpha=0.6)  # Vertical grid lines
    ax.plot(X.T, Y.T, 'k--', linewidth=0.5, alpha=0.6)  # Horizontal grid lines

    # Plot transformed grid
    ax.plot(X_transformed, Y_transformed, 'b', linewidth=1)  # Transformed vertical lines
    ax.plot(X_transformed.T, Y_transformed.T, 'b', linewidth=1)  # Transformed horizontal lines

    # Labels and formatting
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.set_title(f"2D Grid Transformation (Det = {np.linalg.det(A):.2f})")
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)
    ax.set_xlim(-5, 5)
    ax.set_ylim(-5, 5)
    ax.grid(True, linestyle="--", alpha=0.5)

    # Show plot
    plt.show()

# Example 2×2 matrix for transformation
A_2D = np.array([
    [1.5, 0.5],
    [-0.5, 1.5]
])

visualize_grid_transformation(A_2D)
```

## Eigenvalues and eigenvectors
Visualize the eigenvalues and eigenvectors of a matrix using NumPy as stretching factor

```python

def visualize_eigen_stretching_2D(A):
    """
    Visualizes how a 2×2 matrix stretches eigenvectors by plotting transformed unit vectors.
    """
    # Compute eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(A)

    # Define unit vectors
    unit_vectors = np.array([[1, 0], [0, 1]])

    # Apply the matrix transformation
    transformed_vectors = A @ unit_vectors

    # Apply the eigenvalue transformation
    eigen_transformed_vectors = eigenvectors * eigenvalues.reshape(1, -1)

    # Create the plot
    fig, ax = plt.subplots(figsize=(8, 6))

    # Plot unit vectors
    for i in range(unit_vectors.shape[1]):
        ax.quiver(0, 0, unit_vectors[0, i], unit_vectors[1, i], color='black', angles='xy', scale_units='xy', scale=1, label="Original Basis" if i == 0 else "")

    # Plot transformed unit vectors
    for i in range(transformed_vectors.shape[1]):
        ax.quiver(0, 0, transformed_vectors[0, i], transformed_vectors[1, i], color='blue', angles='xy', scale_units='xy', scale=1, label="Transformed Basis" if i == 0 else "")

    # Plot eigenvectors scaled by eigenvalues
    for i in range(eigenvectors.shape[1]):
        ax.quiver(0, 0, eigen_transformed_vectors[0, i], eigen_transformed_vectors[1, i], color='red', angles='xy', scale_units='xy', scale=1, label=f"Eigenvector {i+1} (stretched)")

    # Labels and formatting
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.set_title(f"Eigenvalues as Stretching Factors (2D) (Det = {np.linalg.det(A):.2f})")
    ax.axhline(0, color='black', linewidth=1)
    ax.axvline(0, color='black', linewidth=1)
    ax.set_xlim(-3, 3)
    ax.set_ylim(-3, 3)
    ax.grid(True, linestyle="--", alpha=0.5)
    ax.legend()

    # Show plot
    plt.show()

def visualize_eigen_stretching_3D(A):
    """
    Visualizes how a 3×3 matrix stretches eigenvectors by plotting transformed basis vectors in 3D.
    """
    fig = plt.figure(figsize=(8, 6))
    ax = fig.add_subplot(111, projection='3d')

    # Compute eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(A)

    # Define unit basis vectors in 3D
    unit_vectors = np.eye(3)

    # Apply matrix transformation
    transformed_vectors = A @ unit_vectors

    # Apply eigenvalue transformation
    eigen_transformed_vectors = eigenvectors * eigenvalues.reshape(1, -1)

    # Plot original basis vectors
    for i in range(3):
        ax.quiver(0, 0, 0, unit_vectors[0, i], unit_vectors[1, i], unit_vectors[2, i], color='black', label="Original Basis" if i == 0 else "")

    # Plot transformed basis vectors
    for i in range(3):
        ax.quiver(0, 0, 0, transformed_vectors[0, i], transformed_vectors[1, i], transformed_vectors[2, i], color='blue', label="Transformed Basis" if i == 0 else "")

    # Plot eigenvectors scaled by eigenvalues
    for i in range(3):
        ax.quiver(0, 0, 0, eigen_transformed_vectors[0, i], eigen_transformed_vectors[1, i], eigen_transformed_vectors[2, i], color='red', label=f"Eigenvector {i+1} (stretched)")

    # Labels and formatting
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")
    ax.set_zlabel("Z-axis")
    ax.set_title(f"Eigenvalues as Stretching Factors (3D) (Det = {np.linalg.det(A):.2f})")
    ax.legend()
    ax.grid(True, linestyle="--", alpha=0.5)

    # Show plot
    plt.show()

# Example matrices for 2D and 3D
A_2D = np.array([[2, 1], [1, 2]])
A_3D = np.array([[2, 1, 0], [1, 2, 1], [0, 1, 2]])

# Run both visualizations
visualize_eigen_stretching_2D(A_2D)
visualize_eigen_stretching_3D(A_3D)
```

### Diagonalization of a matrix

```python
def diagonalize_and_visualize(A):
    """
    Computes the diagonalization of a matrix and visualizes
    the original, eigenvector matrix, diagonalized matrix, and reconstructed matrix using heatmaps.
    """
    # Compute eigenvalues and eigenvectors
    eigenvalues, P = np.linalg.eig(A)
    D = np.diag(eigenvalues)  # Diagonal matrix
    P_inv = np.linalg.inv(P)  # Inverse of P
    reconstructed_A = P @ D @ P_inv  # Should be approximately equal to A

    # Create figure for heatmaps
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))

    # Titles for heatmaps
    titles = ["Original Matrix (A)", "Eigenvector Matrix (P)", "Diagonal Matrix (D)", "Reconstructed Matrix (P D P⁻¹)"]
    matrices = [A, P, D, reconstructed_A]

    # Plot heatmaps
    for ax, matrix, title in zip(axes.flat, matrices, titles):
        sns.heatmap(matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5, ax=ax)
        ax.set_title(title)

    plt.tight_layout()
    plt.show()

# Example matrix for diagonalization
A_example = np.array([
    [6, -2, 2],
    [-2, 3, -1],
    [2, -1, 3]
])

diagonalize_and_visualize(A_example)
```

### Solve heat equation using finite difference method
Using the finite difference method to solve the heat equation

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

def solve_heat_equation(dx=0.1, dt=0.01, L=1.0, T=1.0, alpha=0.1):
    """
    Solves the 1D heat equation using the explicit finite difference method.

    Parameters:
    - dx: Spatial step size
    - dt: Time step size
    - L: Length of the rod
    - T: Total time for simulation
    - alpha: Thermal diffusivity

    Returns:
    - A visualization of the heat distribution over time.
    """

    # Discretize space and time
    Nx = int(L / dx) + 1  # Number of spatial points
    Nt = int(T / dt) + 1  # Number of time steps

    # Stability condition for explicit method: dt <= dx^2 / (2*alpha)
    if dt > dx**2 / (2 * alpha):
        raise ValueError("Time step dt is too large! Use a smaller dt for stability.")

    # Initialize temperature distribution: T(x,0) = sin(pi*x) (initial condition)
    x = np.linspace(0, L, Nx)
    u = np.sin(np.pi * x)

    # Boundary conditions: u(0,t) = u(L,t) = 0 (Dirichlet conditions)
    u[0] = 0
    u[-1] = 0

    # Solution matrix (store time evolution)
    U = np.zeros((Nt, Nx))
    U[0, :] = u

    # Time stepping loop (Explicit Finite Difference Method)
    for n in range(0, Nt - 1):
        for i in range(1, Nx - 1):
            U[n + 1, i] = u[i] + alpha * dt / dx**2 * (u[i - 1] - 2 * u[i] + u[i + 1])
        u[:] = U[n + 1, :]  # Update u for the next time step

    # Visualization
    fig, ax = plt.subplots()
    ax.set_xlim(0, L)
    ax.set_ylim(0, 1)
    line, = ax.plot([], [], 'r-', lw=2)

    def update(frame):
        line.set_data(x, U[frame, :])
        return line,

    ani = animation.FuncAnimation(fig, update, frames=Nt, interval=50, blit=True)

    plt.xlabel("Position x")
    plt.ylabel("Temperature u(x,t)")
    plt.title("1D Heat Equation Solution (Explicit Method)")
    plt.show()

# Run the solver
solve_heat_equation()
```

Create the 2D version

```python
def solve_heat_equation_2D(nx=20, ny=20, dx=0.1, dy=0.1, dt=0.01, T=2.0, alpha=0.1):
    """
    Solves the 2D heat equation using the explicit finite difference method.

    Parameters:
    - nx, ny: Number of grid points in x and y directions
    - dx, dy: Spatial step sizes
    - dt: Time step size
    - T: Total time for simulation
    - alpha: Thermal diffusivity

    Returns:
    - A visualization of the heat distribution over time.
    """

    # Stability condition for explicit method: dt <= min(dx^2, dy^2) / (4*alpha)
    if dt > min(dx**2, dy**2) / (4 * alpha):
        raise ValueError("Time step dt is too large! Use a smaller dt for stability.")

    # Grid setup
    x = np.linspace(0, (nx-1)*dx, nx)
    y = np.linspace(0, (ny-1)*dy, ny)
    X, Y = np.meshgrid(x, y)

    # Initial condition: heat in the center
    u = np.zeros((ny, nx))
    u[ny//2, nx//2] = 30  # Heat source in the middle

    # Number of time steps
    Nt = int(T / dt)

    # Solution storage
    U = np.zeros((Nt, ny, nx))
    U[0, :, :] = u

    # Time stepping loop (Explicit Finite Difference Method)
    for n in range(0, Nt - 1):
        u_new = u.copy()
        for i in range(1, ny - 1):
            for j in range(1, nx - 1):
                u_new[i, j] = u[i, j] + alpha * dt * (
                        (u[i+1, j] - 2*u[i, j] + u[i-1, j]) / dx**2 +
                        (u[i, j+1] - 2*u[i, j] + u[i, j-1]) / dy**2
                )
        u[:] = u_new  # Update u for the next time step
        U[n + 1, :, :] = u

    # Visualization
    fig, ax = plt.subplots(figsize=(6,6))
    cax = ax.imshow(U[0], cmap="hot", interpolation="nearest", origin="lower", extent=[0, nx*dx, 0, ny*dy])
    plt.colorbar(cax, label="Temperature")

    def update(frame):
        cax.set_array(U[frame])
        ax.set_title(f"2D Heat Diffusion (Time Step {frame})")
        return cax,

    ani = animation.FuncAnimation(fig, update, frames=Nt, interval=50, blit=False)

    plt.xlabel("X position")
    plt.ylabel("Y position")
    plt.title("2D Heat Equation Solution")
    plt.show()
    return ani

# Run the 2D heat equation solver
ani = solve_heat_equation_2D(alpha=0.15)
```

## PCA
PCA for dimensionality reduction with NumPy

```python
def pca_example():
    """
    Performs Principal Component Analysis (PCA) on a 2D dataset and visualizes the results.
    """

    # Generate synthetic 2D data with correlation
    np.random.seed(42)
    mean = [2, 3]  # Mean of the data
    cov = [[3, 2], [2, 2]]  # Covariance matrix (correlated features)
    X = np.random.multivariate_normal(mean, cov, 100)  # 100 data points

    # Center the data (subtract the mean)
    X_mean = np.mean(X, axis=0)
    X_centered = X - X_mean

    # Compute covariance matrix
    cov_matrix = np.cov(X_centered.T)

    # Compute eigenvalues and eigenvectors
    eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)

    # Sort eigenvectors by descending eigenvalues
    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:, sorted_indices]

    # Transform data to principal component space
    X_pca = X_centered @ eigenvectors

    # Visualization of original data and principal components
    fig, ax = plt.subplots(2, 1, figsize=(5, 12))

    # Original data with eigenvectors (principal components)
    ax[0].scatter(X[:, 0], X[:, 1], alpha=0.6, label="Original Data")
    ax[0].quiver(X_mean[0], X_mean[1], eigenvectors[0, 0], eigenvectors[1, 0], scale=3, color="red", label="PC1")
    ax[0].quiver(X_mean[0], X_mean[1], eigenvectors[0, 1], eigenvectors[1, 1], scale=3, color="blue", label="PC2")
    ax[0].set_title("Original Data with Principal Components")
    ax[0].set_xlabel("Feature 1")
    ax[0].set_ylabel("Feature 2")
    ax[0].legend()
    ax[0].grid(True)

    # Transformed data in PCA space
    ax[1].scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.6, label="PCA Transformed Data", color="green")
    ax[1].axhline(0, color="black", linestyle="--")
    ax[1].axvline(0, color="black", linestyle="--")
    ax[1].set_title("Data Transformed to Principal Components")
    ax[1].set_xlabel("Principal Component 1")
    ax[1].set_ylabel("Principal Component 2")
    ax[1].legend()
    ax[1].grid(True)

    plt.tight_layout()
    plt.show()

# Run PCA example
pca_example()
```

### SVD with NumPy
Perform Singular Value Decomposition (SVD) with NumPy

```python
### SVD

def svd_visualization():
    """
    Performs Singular Value Decomposition (SVD) on a matrix and visualizes
    the original, decomposed components (U, Sigma, V^T), and reconstructed matrix using heatmaps.
    """
    # Create a synthetic 5x5 matrix
    A = np.array([
        [4, 2, 3, 0, 1],
        [2, 3, 1, 5, 4],
        [3, 1, 2, 4, 0],
        [0, 5, 4, 3, 2],
        [1, 4, 0, 2, 3]
    ])

    # Compute SVD: A = U Σ V^T
    U, S, Vt = np.linalg.svd(A)

    # Convert singular values into a diagonal matrix for visualization
    Sigma = np.zeros_like(A, dtype=float)
    np.fill_diagonal(Sigma, S)

    # Reconstruct A using U, Sigma, and V^T
    reconstructed_A = U @ Sigma @ Vt

    # Create figure for heatmaps
    fig, axes = plt.subplots(2, 2, figsize=(12, 10))

    # Titles for heatmaps
    titles = ["Original Matrix (A)", "Left Singular Vectors (U)", "Singular Values (Σ)", "Reconstructed Matrix (U Σ V^T)"]
    matrices = [A, U, Sigma, reconstructed_A]

    # Plot heatmaps
    for ax, matrix, title in zip(axes.flat, matrices, titles):
        sns.heatmap(matrix, annot=True, fmt=".2f", cmap="coolwarm", linewidths=0.5, ax=ax)
        ax.set_title(title)

    plt.tight_layout()
    plt.show()

# Run SVD visualization
svd_visualization()
```
